```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

cities_coordinates = np.array([
    [392.8, 356.4], [559.6, 404.8], [451.6, 186.0], [698.8, 239.6],
    [204.0, 243.2], [590.8, 263.2], [389.2, 448.4], [179.6, 371.2],
    [719.6, 205.2], [489.6, 442.0]
])

G = nx.Graph()

for i in range(len(cities_coordinates)):
    G.add_node(i, pos=(cities_coordinates[i, 0], cities_coordinates[i, 1]))

for i in range(len(cities_coordinates)):
    for j in range(i + 1, len(cities_coordinates)):
        distance = np.linalg.norm(cities_coordinates[i] - cities_coordinates[j])
        G.add_edge(i, j, weight=distance)

pos = nx.get_node_attributes(G, 'pos')
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=8)

labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()
```
